```python
import pandas
from scipy import stats

data = pandas.read_csv("student-mat.csv")
data_yes = data[data["romantic"] == "yes"]
data_no = data[data["romantic"] == "no"]

data_yes.to_csv("data_yes.csv")
data_no.to_csv("data_no.csv")

bins = range(0, 20)
data_yes["G3"].hist(bins=bins)
```

```python
bins = range(0, 20)
data_no["G3"].hist(bins=bins)
```

```python
data_no["G3"].plot.kde()
data_yes["G3"].plot.kde()
```

Pouijeme Mann Whitney test.

- H0: Rozdlen vsledk student se vztahem a student bez vztahu jsou stejn
- H1: Rozdlen vsledk student se vztahem a student bez vztahu jsou rzn

```python
stats.mannwhitneyu(data_yes["G3"], data_no["G3"])
```

p-hodnota je vt ne 5 %. Na hladin vznamnosti 5 % nezamtme H0, tj. neprokzali jsme, e by vsledky tch dvou skupin mly rzn rozdlen.
